```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("world_population.csv")
```

```python
print(df.head())

```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
avg_growth = df.groupby("Continent")["Growth Rate"].mean()
avg_growth.plot(kind="bar", color='skyblue')
plt.title("Average Growth Rate by Continent")
plt.ylabel("Growth Rate (%)")
plt.show()
```

```python
sns.boxplot(x=df["Density (per km²)"])
plt.title("Boxplot of Population Density")
plt.show()
```

```python
numeric_df = df.select_dtypes(include='number')
corr = numeric_df.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
df["Growth Rate"] = df["World Population Percentage"]
plt.hist(df["Growth Rate"], bins=20, edgecolor='black')
plt.title("Histogram of Growth Rate")
plt.xlabel("Growth Rate (%)")
plt.ylabel("Number of Countries")
plt.show()

```

```python
import pandas as pd

df = pd.read_csv("World_population.csv")
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Select only numeric columns
numeric_df = df.select_dtypes(include=np.number)

# Compute correlation matrix
corr = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
numeric_df = df.select_dtypes(include='number')
corr = numeric_df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
import numpy as np

mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", mask=mask)
```

```python
plt.savefig('correlation_heatmap.png', dpi=300, bbox_inches='tight')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("world_population.csv")

# Group by continent and sum 2022 population
continent_pop = df.groupby("Continent")["2022 Population"].sum()

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    continent_pop, 
    labels=continent_pop.index, 
    autopct='%1.1f%%', 
    startangle=140
)
plt.title("World Population Distribution by Continent (2022)")
plt.show()
```
